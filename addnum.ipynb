```python
#libary to store and data manipulate
import pandas as pd

#read the data frame and asssign that data set to a name as data to reuse
data=pd.read_csv('AI_New_1.csv')

#gain the data set named as data above
data

#output the first values of the data set
data.head()

#output the last values of the data set
data.tail()

#gain the number of rows and colums in the data set
data.shape

#check the whether any missing values are there
data.info()

#libary to analyze the combination between inputs and output
import matplotlib.pyplot as plt

#gain a scatter plot fo x and sum
plt.scatter(data['x'],data['sum'])

plt.scatter(data['y'],data['sum'])

X=data[['x','y']]
Y=data['sum']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

X_train

from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(X_train,y_train)

model.score(X_train,y_train)

model.score(X_test,y_test)

y_pred=model.predict(X_test)

y_pred

#compare it with actual value
y_test

df=pd.DataFrame({'Actual':y_test,'Prediction':y_pred})

#compae using a dataframe
df

#we can show the predictions using saving the model
import joblib

#model will be saved to the below name
joblib.dump(model,'model_joblib')

#load the model
model=joblib.load('model_joblib')

#prediction
model.predict([[23,45]])

#train entire dataset
X=data[['x','y']]
y=data['sum']

model=LinearRegression()
model.fit(X,y)

import joblib
joblib.dump(model,'model_1.pkl')
model=joblib.load('model_1.pkl')
model.predict([[23,45]])

```

```python

# Define a function to handle button click event
def show_entry_fields():
    # Retrieve the values entered in the Entry widgets and convert them to float
    p1 = float(e1.get())
    p2 = float(e2.get())

    # Load the machine learning model stored in a file named 'model_1' using joblib
    model = joblib.load('model_1.pkl')

    # Make a prediction using the loaded model and the input values
    result = model.predict([[p1, p2]])

    # Display the prediction result on the Tkinter window
    Label(master, text='Sum is - ').grid(row=4)  # Label for displaying text 'Sum is - ' on row 4
    Label(master, text=result).grid(row=5)        # Label for displaying the prediction result on row 5

    # Print the prediction result to the console
    print("Sum is - ", result)
```

```python
from tkinter import *

# Import the necessary module(s) from tkinter library
import joblib

# Create the main Tkinter window
master = Tk()

master.title("Addition of Two Numbers Using ML")
Label(master, text="Addition of two numbers using ML", bg='black', fg='white').grid(row=0, columnspan=2)

# Create and place labels for indicating where to enter the first and second values
Label(master, text="Enter 1st").grid(row=1)  # Label for indicating where to enter the first value on row 1
Label(master, text="Enter 2nd").grid(row=2)  # Label for indicating where to enter the second value on row 2

# Create Entry widgets for entering the first and second values, and place them in the window
e1 = Entry(master)  # Entry widget for entering the first value
e2 = Entry(master)  # Entry widget for entering the second value

# Place the Entry widgets in the window
e1.grid(row=1, column=1)  # Place the first Entry widget in the first column of row 1
e2.grid(row=2, column=1)  # Place the second Entry widget in the first column of row 2

# Create a button for triggering the prediction process and place it in the window
Button(master, text='Predict', command=show_entry_fields).grid(row=3, columnspan=2)  # Button for triggering the prediction process

# Start the Tkinter event loop
mainloop()


```
